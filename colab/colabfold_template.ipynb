# ColabFold-Workflow für das KI-Strukturmodell-Labor

Dieses Notebook ist **nicht** die eigentliche Vergleichs-Webseite. Es ist die externe KI-Vertiefung:

> Sequenz vorbereiten → in ColabFold falten → PDB herunterladen → im Webtool importieren → bei Bedarf ins Repo übernehmen.

Die Website bleibt bewusst stabil und führt keine Live-Faltung aus.


## Ablauf in Kurzform

1. Beispiel auswählen (`selected = 'ubiquitin'` oder `selected = 'trp_cage'`).
2. Zellen bis zum FASTA-Block ausführen.
3. Den angezeigten FASTA-Block kopieren.
4. Das öffentliche ColabFold/AlphaFold2-Notebook öffnen.
5. FASTA dort einfügen und Modell berechnen lassen.
6. PDB-Datei herunterladen.
7. PDB zuerst im Webtool über **Eigenes PDB testen** importieren.
8. Erst wenn das Modell als Standardvergleich geeignet ist, Datei ins Repo übernehmen.


````python
examples = {
    'trp_cage': {
        'title': 'Trp-cage',
        'sequence': 'NLYIQWLKDGGPSSGRPPPS',
        'target_path': 'structures/trp_cage/af2_colabfold.pdb',
        'didactic_note': 'Mini-Protein: kurze Sequenz, bewegliches System, guter Einstieg.'
    },
    'ubiquitin': {
        'title': 'Ubiquitin',
        'sequence': 'MQIFVKTLTGKTITLEVEPSDTIENVKAKIQDKEGIPPDQQRLIFAGKQLEDGRTLSDYNIQKESTLHLVLRLRGG',
        'target_path': 'structures/ubiquitin/af2_colabfold.pdb',
        'didactic_note': 'Referenzprotein: stabiler Fold, sehr guter Vergleich KI-Modell vs. Experiment.'
    }
}

print('Verfügbare Beispiele:')
for key, item in examples.items():
    print(f'- {key}: {item["title"]} → {item["didactic_note"]}')
````

## 1. Beispiel auswählen

Ändere bei Bedarf nur die folgende Variable. Für den ersten Standardvergleich ist `ubiquitin` ideal.


````python
selected = 'ubiquitin'  # 'ubiquitin' oder 'trp_cage'
item = examples[selected]
print('Ausgewählt:', item['title'])
print('Sequenzlänge:', len(item['sequence']), 'Aminosäuren')
print('Didaktischer Hinweis:', item['didactic_note'])
print('Zielpfad im Repo:', item['target_path'])
````

## 2. FASTA-Block zum Kopieren

Kopiere den folgenden FASTA-Block in das externe ColabFold/AlphaFold2-Notebook.
Die automatisch geschriebene `.fasta`-Datei ist nur eine Hilfe im aktuellen Colab-Dateisystem; sie wird nicht automatisch an ColabFold übertragen.


````python
fasta_text = f">{item['title']}\n{item['sequence']}\n"
print('--- FASTA-BLOCK ZUM KOPIEREN ---')
print(fasta_text)
print('--- ZIELPFAD FÜR DAS KURATIERTE STANDARDMODELL ---')
print(item['target_path'])
with open(f"{selected}.fasta", 'w', encoding='utf-8') as f:
    f.write(fasta_text)
print('FASTA-Datei zusätzlich im aktuellen Colab-Dateisystem geschrieben:', f'{selected}.fasta')
````

## 3. Externes ColabFold/AlphaFold2-Notebook öffnen

Öffne das öffentliche ColabFold-Notebook:

[ColabFold AlphaFold2 Notebook öffnen](https://colab.research.google.com/github/sokrypton/ColabFold/blob/main/AlphaFold2.ipynb)

Dort den FASTA-Block einfügen. Für diese einfachen Beispiele reichen zunächst die Standardparameter.

Für spätere didaktische Vergleiche kann man nicht nur das beste Modell, sondern auch schlechter bewertete Modelle herunterladen und im Webtool testen.


## 4. Ergebnis zurück ins Webtool

### Variante A: zuerst testen

PDB-Datei aus ColabFold herunterladen und im Webtool über **Eigenes PDB testen** importieren.

### Variante B: als Standardmodell ins Repo übernehmen

Wenn das Modell geeignet ist, die Datei exakt umbenennen und ablegen als:

```text
structures/ubiquitin/af2_colabfold.pdb
structures/trp_cage/af2_colabfold.pdb
```

Danach GitHub Pages kurz deployen lassen und die Webseite mit Strg+F5 neu laden.
